# Projeto Spaceship Titanic

## Contexto

Este projeto tem por função a partir de um cenário ficticio onde uma nave está transportando diversos passageiros prever quais destes passageiros foram transportados para uma dimensão alternativa com base nos dados de cada passageiro, isto deve ser feito treinando um modelo e a partir deste modelo tentando realizar tal previsão posteriormente.

## Objetivo

Colocar em pratica um modelo de previsão funcional com base em uma situação fictícia proposta.


## Créditos 

As bases de dados foram retiradas do site kaggle: https://www.kaggle.com/competitions/spaceship-titanic/overview

Os dados estão dispostos nas seguintes colunas:

- PassengerId - Um Id único para cada passageiro. Cada Id segue o formato gggg_pp, onde gggg indica um grupo com o qual o passageiro está viajando e pp é o número dele dentro do grupo. As pessoas em um grupo geralmente são membros da família, mas nem sempre.
 
- HomePlanet - O planeta de onde o passageiro partiu, normalmente o planeta de residência permanente.

- CryoSleep - Indica se o passageiro optou por entrar em animação suspensa durante a duração da viagem. Os passageiros em criossono estão confinados às suas cabines.

- Cabin - O número da cabine onde o passageiro está hospedado. Segue o formato deck/num/side, onde side pode ser P para boreste ou S para estibordo.

- Destination - O planeta para o qual o passageiro vai desembarcar.

- Age - A idade do passageiro.

- VIP - Se o passageiro pagou por um serviço VIP especial durante a viagem.
 
- RoomService, FoodCourt, ShoppingMall, Spa, VRDeck - Valor que o passageiro gastou em cada uma das várias comodidades de luxo da Spaceship Titanic.

- Name - O nome e sobrenome do passageiro.

- Transported - Se o passageiro foi transportado para outra dimensão. Esse é o alvo, a coluna que você está tentando prever.

## Declarando funções úteis

```python
def limites(coluna):
    q1 = coluna.quantile(0.25)
    q3 = coluna.quantile(0.75)
    amplitude = q3 - q1
    return q1 - 1.5*amplitude, q3 + 1.5*amplitude

def excluir_outliers(df,nome_coluna):
    n_linhas = df.shape[0]
    lim_inferior, lim_superior = limites(df[nome_coluna])
    df = df.loc[(df[nome_coluna] >= lim_inferior) & (df[nome_coluna] <= lim_superior) , :]
    linhas_removidas = n_linhas - df.shape[0]
    return df, linhas_removidas
```

```python
def diagrama_caixa(coluna):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(15,5)
    sns.boxplot(x = coluna, ax = ax1)
    ax2.set_xlim(limites(coluna))
    sns.boxplot(x = coluna,ax = ax2)

def histograma(coluna):
    plt.figure(figsize = (20, 10))
    sns.histplot(data =coluna,bins = 35, kde=True)
    
def grafico_barras(coluna):
    plt.figure(figsize = (15,5))
    ax = sns.barplot(x =coluna.value_counts().index ,y = coluna.value_counts())
```

```python
fig = plt.figure(figsize = (10,8))
sns.heatmap(data_df.corr(), annot = True, cmap = "Greens")
```

## Modelagem

Inicialmente serão criadas colunas dummies para poder aplicar os modelos e serão separados os dataframes de teste e treino de modelo.

```python
colunas_dummies = ["HomePlanet", "Deck", "Side", "Destination"]
data_df = pd.get_dummies(data = data_df, columns = colunas_dummies)

display(data_df.head())
```

```python
treino_final_df = data_df.loc[(data_df["Transported"] == 1) | (data_df["Transported"] == 0)]
teste_final_df = data_df.loc[data_df["Transported"] == 2]

teste_final_df = teste_final_df.drop(columns = "Transported")

display(teste_final_df)
display(treino_final_df)
```

```python
display(teste_final_df.isnull().sum())

colunas_fillna0 = ["Num", "CryoSleep", "VIP"]

colunas_numeros =  ["Age", "RoomService","FoodCourt", "ShoppingMall", "Spa", "VRDeck"]

for coluna in colunas_numeros:
    teste_final_df[coluna] = teste_final_df[coluna].fillna(data_df[coluna].mean())

for coluna in colunas_fillna0:
    teste_final_df[coluna] = teste_final_df[coluna].fillna(0)
```

```python
display(teste_final_df.isnull().sum())

display(teste_final_df)
```

### Modelos de previsão usados 

Serão usados os modelos de:
- Linear Regression
- Random Forest
- Extra Trees

Serão divididos o dataframe de treino do modelo na coluna "Tranported" e nas demais colunas.

```python
modelo_lr = LogisticRegression()
modelo_rf = RandomForestClassifier()
modelo_kn = KNeighborsClassifier()
modelo_gn = GaussianNB()

modelos = {"LogisticRegression": modelo_lr,
          "RandomForest": modelo_rf,
          "KNeighbors": modelo_kn,
          "GaussianNB": modelo_gn}

X_treino = treino_final_df.drop(columns = "Transported").astype('int')
y_treino = treino_final_df["Transported"].astype('int')

display(X_treino)
display(y_treino)
```

### K-Fold Cross Validation:

Agora usaremos o K-Fold Cross Validation para estimar qual seria o modelo com a maior precisão inicialmnete depois tentando refinar ainda mais o modelo para obter uma maior precisão:

```python
for nome_modelo, modelo in modelos.items():
    inicio = time.time()
    scores = cross_val_score(modelo, X_treino, y_treino, cv=10, scoring = "accuracy")
    print("Tempo necessário para modelo {} :{:.2f}s".format(nome_modelo, (time.time() - inicio)))
    print("Score:", scores)
    print("Média:", scores.mean())
    print("Desvio Padrão:", scores.std())
    print("\n")
```

  Como pode-se notar o modelo Random Forest tendo a maior precisão entre todos os modelos e um desvio padrão consideravelmente baixo (3,4%).
  
  Agora analisaremos qual a importancia de cada parametro dentro do modelo para obter uma maior precisão se possível.

```python
modelo_rf.fit(X_treino, y_treino)
parametros = pd.DataFrame({'feature':X_treino.columns,'Parametros':np.round(modelo_rf.feature_importances_,3)})
parametros = parametros.sort_values('Parametros',ascending=False).set_index('feature')
display(parametros)
parametros.plot.bar()
```

Nota-se que a coluna VIP e as colunas relacionadas ao destino dos passageiros não tem tanta importaância dentro do modelo, assim elas serão removidas e o modelo será testado novamente para testar sua precisão.

```python
X_treino_novo = X_treino.drop(columns = ['Destination_PSO J318.5-22','Destination_TRAPPIST-1e', 'Destination_55 Cancri e', "VIP"])
```

```python
scores = cross_val_score(modelo_rf, X_treino_novo, y_treino, cv=10, scoring = "accuracy")
print("Score:", scores)
print("Média:", scores.mean())
print("Desvio Padrão:", scores.std())
```

  Apesar dessa mudança ter deixado o modelo mais simples observou-se um aumento no desvio padrão e uma leve queda de precisão assim optou-se por manter o modelo com todas as colunas.

Finalmente será usado o modelo para previsão com a criação de um arquivo com as previsões finais usando o modelo Random Forest.

```python
inicio = time.time()
#treinando o modelo
modelo_rf.fit(X_treino, y_treino)
#testando o modelo:
previsao = modelo_rf.predict(teste_final_df)
print("Tempo necessário para modelo Random Forest :{:.2f}s".format(time.time() - inicio))
previsao_final = pd.DataFrame(list(zip(teste_final_df["PassengerId"], previsao)), columns = ["PassengerId", "Transported"])
previsao_final = previsao_final.set_index("PassengerId")
display(previsao_final)
previsao_final.to_excel('submission.xlsx')
```
